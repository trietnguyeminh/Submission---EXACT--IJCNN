```python
# Cài đặt các thư viện cần thiết (chạy 1 lần)
!pip install -q transformers evaluate rouge_score bert_score

import os
import zipfile
import shutil
import glob
import csv
import torch
import torch.nn as nn
import pandas as pd
from torch.utils.data import Dataset, DataLoader
from torch.cuda.amp import autocast, GradScaler
from torchvision import transforms
from PIL import Image
from tqdm import tqdm
from transformers import AutoTokenizer, ViTModel, GPT2LMHeadModel

# Bật GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f" Đang chạy trên thiết bị: {device}")
```

```python
def extract_zip(zip_path, extract_to):
    print(f" Đang giải nén {zip_path}...")
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_to)
    print(f" Đã giải nén vào {extract_to}")

# 1. GIẢI NÉN VÀ CHIA ẢNH TRAIN/VALID (Code gốc của bạn)
def unzip_and_split_dataset(zip_path, base_extract_dir='/content/dataset'):
    extract_zip(zip_path, base_extract_dir)
    train_dir = os.path.join(base_extract_dir, 'train_images')
    valid_dir = os.path.join(base_extract_dir, 'valid_images')
    os.makedirs(train_dir, exist_ok=True)
    os.makedirs(valid_dir, exist_ok=True)

    count_train, count_valid = 0, 0
    for root, dirs, files in os.walk(base_extract_dir):
        if root in [train_dir, valid_dir]: continue
        for file in files:
            if file.lower().endswith(('.png', '.jpg', '.jpeg', '.tif', '.bmp')):
                file_path = os.path.join(root, file)
                if 'train' in file.lower():
                    shutil.move(file_path, os.path.join(train_dir, file))
                    count_train += 1
                elif 'valid' in file.lower():
                    shutil.move(file_path, os.path.join(valid_dir, file))
                    count_valid += 1

    print(f" Hoàn thành: {count_train} ảnh train | {count_valid} ảnh valid")
    return train_dir, valid_dir

# Chạy giải nén ảnh gốc
images_zip = "/content/drive/MyDrive/Colab Notebooks/images.zip"
TRAIN_IMG_DIR, VALID_IMG_DIR = unzip_and_split_dataset(images_zip)

# 2. GIẢI NÉN BỘ TEST MỚI (Dành cho nộp bài 1/5)
test_zip = "/content/drive/MyDrive/Colab Notebooks/test_synth.zip"
TEST_IMG_DIR = '/content/dataset/test_images'
extract_zip(test_zip, TEST_IMG_DIR)

# 3. GIẢI NÉN BỘ CONCEPTS & TỪ ĐIỂN
cui_zip = "/content/drive/MyDrive/Colab Notebooks/CLEF 2026 Supplements/CLEF 2026 Supplements/cui_inventory_names.zip"
concepts_zip = "/content/drive/MyDrive/Colab Notebooks/CLEF 2026 Supplements/CLEF 2026 Supplements/manuel_concepts.zip"
SUPP_DIR = '/content/supplements'

extract_zip(cui_zip, SUPP_DIR)
extract_zip(concepts_zip, SUPP_DIR)
```

```python
print("Đang nạp từ điển bệnh lý và Caption...")

# 1. Đọc số lượng Concepts (Làm não bộ cho XAI)
# Tìm file csv trong thư mục giải nén
cui_file = glob.glob(f"{SUPP_DIR}/**/*.csv", recursive=True)[0]
cui_vocab_df = pd.read_csv(cui_file)
REAL_CONCEPTS = cui_vocab_df['CUI'].tolist() if 'CUI' in cui_vocab_df.columns else cui_vocab_df.iloc[:, 0].tolist()
NUM_CONCEPTS = len(REAL_CONCEPTS)
print(f" Tổng số loại bệnh lý (Concepts) mô hình cần học: {NUM_CONCEPTS}")

# 2. Mapping Concepts cho từng ảnh
concept_dict = {}
concept_files = glob.glob(f"{SUPP_DIR}/**/train*.csv", recursive=True) + glob.glob(f"{SUPP_DIR}/**/valid*.csv", recursive=True)

for file in concept_files:
    df_concept = pd.read_csv(file)
    for _, row in df_concept.iterrows():
        cuis = str(row['CUIs']).split(';') if pd.notna(row['CUIs']) else []
        concept_dict[row['ID']] = cuis

# 3. Đọc Captions
CAPTIONS_CSV = '/content/drive/MyDrive/Colab Notebooks/captions.csv'
df_all = pd.read_csv(CAPTIONS_CSV)

# Lọc file theo tên (Giả định ID chứa chữ train/valid)
train_df = df_all[df_all['ID'].str.contains('train', case=False)].reset_index(drop=True)
valid_df = df_all[df_all['ID'].str.contains('valid', case=False)].reset_index(drop=True)
```

```python
# QUAN TRỌNG: padding_side="left" để sinh câu theo Batch không bị lỗi
tokenizer = AutoTokenizer.from_pretrained("distilgpt2")
tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = "left"

basic_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

class MultiTaskMedicalDataset(Dataset):
    def __init__(self, df, img_dir, tokenizer, concept_dict, transform=None):
        self.df = df
        self.img_dir = img_dir
        self.tokenizer = tokenizer
        self.concept_dict = concept_dict
        self.transform = transform

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        img_id = self.df.loc[idx, 'ID']
        caption = self.df.loc[idx, 'Caption']

        # Dùng glob để tìm ảnh chính xác (bất chấp đuôi jpg hay png)
        img_search = glob.glob(os.path.join(self.img_dir, f"{img_id}.*"))
        img_path = img_search[0] if img_search else os.path.join(self.img_dir, f"{img_id}.jpg")

        try:
            image = Image.open(img_path).convert('RGB')
        except:
            image = Image.new('RGB', (224, 224))

        if self.transform: image = self.transform(image)

        encoded_caption = self.tokenizer(
            caption, padding='max_length', truncation=True, max_length=64, return_tensors="pt"
        )

        true_cuis = self.concept_dict.get(img_id, [])
        labels = [1.0 if concept in true_cuis else 0.0 for concept in REAL_CONCEPTS]

        return {
            "img_id": img_id, # Trả về ID để lát xuất file nộp bài
            "pixel_values": image,
            "input_ids": encoded_caption.input_ids.squeeze(),
            "attention_mask": encoded_caption.attention_mask.squeeze(),
            "concept_labels": torch.tensor(labels, dtype=torch.float)
        }

# pin_memory=True giúp đẩy data lên GPU T4 nhanh hơn
train_dataset = MultiTaskMedicalDataset(train_df, TRAIN_IMG_DIR, tokenizer, concept_dict, basic_transform)
valid_dataset = MultiTaskMedicalDataset(valid_df, VALID_IMG_DIR, tokenizer, concept_dict, basic_transform)

train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True, num_workers=2, pin_memory=True)
valid_loader = DataLoader(valid_dataset, batch_size=32, shuffle=False, num_workers=2, pin_memory=True)
```

```python
class MedicalCaptioningMultiTaskModel(nn.Module):
    def __init__(self, num_concepts, vocab_size):
        super().__init__()
        self.encoder = ViTModel.from_pretrained("google/vit-base-patch16-224-in21k")
        vision_dim = self.encoder.config.hidden_size

        self.concept_head = nn.Sequential(
            nn.Linear(vision_dim, 512),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(512, num_concepts)
        )

        self.decoder = GPT2LMHeadModel.from_pretrained("distilgpt2")
        text_dim = self.decoder.config.n_embd
        self.bridge = nn.Linear(vision_dim, text_dim)

    def forward(self, pixel_values, input_ids=None, attention_mask=None):
        encoder_outputs = self.encoder(pixel_values=pixel_values)
        image_features = encoder_outputs.last_hidden_state[:, 0, :]
        concept_logits = self.concept_head(image_features)
        projected_image = self.bridge(image_features).unsqueeze(1)

        text_loss = None
        if input_ids is not None:
            input_embeds = self.decoder.transformer.wte(input_ids)
            combined_embeds = torch.cat((projected_image, input_embeds[:, :-1, :]), dim=1)
            outputs = self.decoder(inputs_embeds=combined_embeds, attention_mask=attention_mask, labels=input_ids)
            text_loss = outputs.loss

        return text_loss, concept_logits

model = MedicalCaptioningMultiTaskModel(NUM_CONCEPTS, tokenizer.vocab_size).to(device)
```

```python
from torch.cuda.amp import autocast, GradScaler
import os

optimizer = torch.optim.AdamW(model.parameters(), lr=5e-5)
concept_loss_fn = nn.BCEWithLogitsLoss()
scaler = GradScaler() # Vũ khí tối mật tăng tốc độ AMP

EPOCHS = 5
CONCEPT_WEIGHT = 1.0
SAVE_DIR = '/content/drive/MyDrive/Colab Notebooks/Weights'
os.makedirs(SAVE_DIR, exist_ok=True)

# File lưu toàn bộ trạng thái để khôi phục
LATEST_CHECKPOINT = os.path.join(SAVE_DIR, 'latest_checkpoint.pt')
start_epoch = 0

# ==========================================
# CƠ CHẾ AUTO-RESUME (TỰ NẠP LẠI KHI BỊ SẬP)
# ==========================================
if os.path.exists(LATEST_CHECKPOINT):
    print("🔄 Phát hiện Checkpoint cũ! Đang khôi phục trí nhớ...")
    checkpoint = torch.load(LATEST_CHECKPOINT, map_location=device)
    model.load_state_dict(checkpoint['model_state_dict'])
    optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
    scaler.load_state_dict(checkpoint['scaler_state_dict'])
    start_epoch = checkpoint['epoch'] + 1
    print(f" Đã nạp thành công! Tiếp tục chạy từ Epoch {start_epoch + 1}")
else:
    print(" BẮT ĐẦU HUẤN LUYỆN TỪ SỐ 0 VỚI MIXED PRECISION...")

# ==========================================
# VÒNG LẶP HUẤN LUYỆN
# ==========================================
for epoch in range(start_epoch, EPOCHS):
    model.train()
    total_loss, total_text, total_con = 0, 0, 0
    progress_bar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{EPOCHS}")

    for batch in progress_bar:
        optimizer.zero_grad()

        pixel_values = batch["pixel_values"].to(device, non_blocking=True)
        input_ids = batch["input_ids"].to(device, non_blocking=True)
        attention_mask = batch["attention_mask"].to(device, non_blocking=True)
        concept_labels = batch["concept_labels"].to(device, non_blocking=True)

        with autocast():
            text_loss, concept_logits = model(pixel_values, input_ids, attention_mask)
            concept_loss = concept_loss_fn(concept_logits, concept_labels)
            loss = text_loss + (CONCEPT_WEIGHT * concept_loss)

        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()

        total_loss += loss.item()
        progress_bar.set_postfix({'Loss': f"{loss.item():.3f}"})

    print(f" Hết Epoch {epoch+1} | Avg Loss: {total_loss/len(train_loader):.4f}")

    # 1. Lưu trọng số xịn để sau này dự đoán
    torch.save(model.state_dict(), os.path.join(SAVE_DIR, f'model_epoch_{epoch+1}.pt'))

    # 2. Lưu trạng thái Checkpoint để chống sập (Ghi đè liên tục)
    torch.save({
        'epoch': epoch,
        'model_state_dict': model.state_dict(),
        'optimizer_state_dict': optimizer.state_dict(),
        'scaler_state_dict': scaler.state_dict()
    }, LATEST_CHECKPOINT)
```
